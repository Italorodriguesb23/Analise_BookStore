```python
import pandas as pd
import seaborn as srn
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dados_limpos.csv')
df.head()
```

```python
agrupado = df['Categoria'].value_counts()
agrupado
```

```python
#Gráfico de barras
top5 = agrupado.nlargest(5) 

top5.plot(kind='bar', color=['blue', 'yellow', 'green', 'red', 'pink'], figsize=(8,5))
plt.title('N° de livros por categoria')
plt.xlabel('Categoria')
plt.ylabel('Quantidade disponível')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
top5_categorias = df['Categoria'].value_counts().head(5).index
```

```python
df_top5 = df[df['Categoria'].isin(top5_categorias)]
```

```python
#Gráfico Boxplot
plt.figure(figsize=(10, 6))
srn.boxplot(x='Categoria', y='Preço', data=df_top5, hue='Categoria', palette='pastel', legend=False)
plt.title('Preços por categoria')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Gráfico Histograma
plt.hist(df['Preço'], bins=6, color = 'skyblue')
plt.title('Distribuição de preços de livros')
plt.xlabel('Livros')
plt.ylabel('Preço')
```

```python
#Gráfico dispersão
srn.scatterplot(data=df, x='Avaliação (1-5)', y='Preço')
```

```python
#Gráfico de pizza
top5.plot.pie(legend=True)
plt.title('Grafico de pizza das categorias')
```

```python
#Gráfico de dispersão com texto
plt.scatter(df['Avaliação (1-5)'], df['Preço'])
plt.axhline(50, color='red', linestyle='--')
plt.text(0, 52, 'Preço acima da média', color='red')
```
